```python
# Generate a sine wave that exactly reaches 1 and -1 at each sample point
num_points = 10  # Number of points
t = np.linspace(0, num_points, num_points, endpoint=False)  # Time array
print(t)
# Calculate the sine wave values
sine_wave_exact = np.sin((np.pi)*(t+0.5))  # 500 Hz to ensure crossing at every step
sine_wave_exact_2 = np.sin((np.pi/2)*(t*0.5) )
sine_wave_exact_4 = np.sin((np.pi/4)*(t+0.5) )
# Plotting the exact sine wave
plt.figure(figsize=(10, 5))
plt.plot(t, sine_wave_exact, label='Exact Sine Wave')
plt.plot(t, sine_wave_exact_2, label='/2')
plt.plot(t, sine_wave_exact_4, label='/4')
plt.title('Exact Sine Wave Crossing -1 and 1 at Every Sample')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axhline(1, color='red', linestyle='--', label='y=1')
plt.axhline(-1, color='blue', linestyle='--', label='y=-1')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to get position encoding
def getPositionEncoding(seq_len, d, n=10000):
    P = np.zeros((seq_len, d))
    for k in range(seq_len):
        for i in np.arange(int(d / 2)):
            denominator = np.power(n, 2 * i / d)
            P[k, 2 * i] = np.sin(k / denominator)
            P[k, 2 * i + 1] = np.cos(k / denominator)
    return P

# Get position encoding for the given parameters
P = getPositionEncoding(seq_len=1000, d=2)

```
