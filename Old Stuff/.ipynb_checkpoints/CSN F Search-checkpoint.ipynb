```python
from csn3 import SineWaveLayer,SelectorWaveLayer,SineWaveNetwork
import torch
import torch.nn as nn

# Function to print phi and w parameters for each layer
def print_parameters(network):
    for name, module in network.named_modules():
        if isinstance(module, (SineWaveLayer, SelectorWaveLayer)):
            print(f"Layer: {name}")
            print(f"  phi: {module.phi.data.shape}")
            print(f"  phic: {module.phi_current.data}")
            print(f"  w: {module.w.data.shape}")
            print(f"  wc: {module.current_w.data}")
            print()

```

```python
class SineWaveNetworkExp(nn.Module):
    def __init__(self):
        super(SineWaveNetworkExp, self).__init__()
        
        # Layers
        self.sine_layer = SineWaveLayer(n=10)
        
        # Selector layers with increasing m
        self.selector_layer1 = SelectorWaveLayer(n=10, m=10, phi_prev=True, w_prev=True)
        self.selector_layer2 = SelectorWaveLayer(n=10, m=80, phi_prev=True, w_prev=True)
        self.selector_layer3 = SelectorWaveLayer(n=80, m=60, phi_prev=True, w_prev=True)
        self.selector_layer4 = SelectorWaveLayer(n=60, m=40, phi_prev=True, w_prev=True)
        self.selector_layer5 = SelectorWaveLayer(n=40, m=20, phi_prev=True, w_prev=True)
        self.selector_layer6 = SelectorWaveLayer(n=20, m=1, phi_prev=True, w_prev=True)

    def forward(self, t_vector, initial_phi=None, initial_w=None):
        # First layer
        sine_output, phi, w = self.sine_layer(t_vector, initial_phi, initial_w)
        
        # Selector layers
        selector_output1, phi, w = self.selector_layer1(t_vector, sine_output, phi, w)
        selector_output2, phi, w = self.selector_layer2(t_vector, selector_output1, phi, w)
        selector_output3, phi, w = self.selector_layer3(t_vector, selector_output2, phi, w)
        selector_output4, phi, w = self.selector_layer4(t_vector, selector_output3, phi, w)
        selector_output5, phi, w = self.selector_layer5(t_vector, selector_output4, phi, w)
        selector_output6, phi, w = self.selector_layer6(t_vector, selector_output5, phi, w)
        
        return selector_output6
    def top_up(self):
        for name, module in self.named_modules():
            if isinstance(module, (SineWaveLayer, SelectorWaveLayer)): 
                module.top_up_phi()
                module.top_up_w()
```

```python
## averages stereo to make mono
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

# Read the .wav file
sample_rate, data = wavfile.read('dog_bark.wav')

# Check if the audio is stereo
if data.ndim == 2:
    # Average the two channels to get a mono signal
    data = data.mean(axis=1)

# Convert the numpy array to a PyTorch tensor
waveform = torch.from_numpy(data).float()

# Plot the waveform
plt.figure(figsize=(12, 6))
plt.plot(waveform.numpy())
plt.title('Waveform of the .wav file (Merged Channels)')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.show()
```

```python
import torch.nn.functional as F

class CosineSimilarityLoss(nn.Module):
    def __init__(self):
        super(CosineSimilarityLoss, self).__init__()
        self.cosine_similarity = nn.CosineSimilarity(dim=1)

    def forward(self, input, target):
        # Calculate cosine similarity
        cosine_sim = self.cosine_similarity(input, target)
        # Cosine similarity ranges from -1 to 1, so to turn it into a loss we do (1 - cosine_similarity)
        return 1 - cosine_sim.mean()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Define hyperparameters
hyperparameters = {
    'learning_rate': 0.0001,
    'epochs': 100,
    'batch_size': 1000,
}

# Example waveform data
wav_data = waveform
time_steps = torch.arange(1, len(wav_data) + 1)

# Define the dataset and dataloader
class WaveformDataset(Dataset):
    def __init__(self, wav_data, time_steps):
        self.wav_data = wav_data
        self.time_steps = time_steps
        self.length = len(wav_data)

    def __len__(self):
        return self.length

    def __getitem__(self, idx):
        return self.wav_data[idx], self.time_steps[idx]

# Create the dataset
wav_data = waveform
time_steps = torch.arange(1, len(wav_data) + 1)
dataset = WaveformDataset(wav_data, time_steps)

# Custom sampler to generate random batches
class RandomBatchSampler(torch.utils.data.Sampler):
    def __init__(self, data_source, batch_size):
        self.data_source = data_source
        self.batch_size = batch_size

    def __iter__(self):
        indices = torch.randperm(len(self.data_source)).tolist()
        for i in range(0, len(indices), self.batch_size):
            yield indices[i:i + self.batch_size]

    def __len__(self):
        return len(self.data_source) // self.batch_size

# Create the dataloader with the custom random batch sampler
batch_size = hyperparameters['batch_size']
sampler = RandomBatchSampler(dataset, batch_size)
dataloader = DataLoader(dataset, batch_sampler=sampler)
# Instantiate the network
network = SineWaveNetworkExp()

# Define the loss function and optimizer
criterion = nn.MSELoss()
#criterion = CosineSimilarityLoss()
optimizer = optim.Adam(network.parameters(), lr=hyperparameters['learning_rate'])

# Training loop
for epoch in range(hyperparameters['epochs']):
    running_loss = 0.0
    for i, (batch_wav, batch_time) in enumerate(dataloader):

        
        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        output = network(batch_time)
        ##print(batch_time.shape,batch_wav.shape, output.shape )
   
        # Calculate loss
        loss = criterion(output, batch_wav)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()
        network.top_up()
        #i think right here i add my adjust function? 

        # Print statistics
        running_loss += loss.item()
    print(epoch,running_loss)

print('Finished Training')
```

```python
import matplotlib.pyplot as plt
import torch

def increment_and_graph(network, time_steps, delta, n):
    # List to store outputs for each iteration
    outputs = []

    # Set the network to evaluation mode
    network.eval()

    # Perform the increments and collect the outputs
    for _ in range(n):
        # Increment current_w by delta for each layer
        for module in network.modules():
            if isinstance(module, (SineWaveLayer, SelectorWaveLayer)):
                module.current_w.data += delta

        # Get the network output
        pred = network(time_steps).detach().numpy()
        outputs.append(pred)

    # Plotting the waveforms with different colors
    plt.figure(figsize=(10, 4))
    for i, output in enumerate(outputs):
        plt.plot(output.squeeze(0), label=f'Iteration {i + 1}')

    plt.title("Generated Waveforms")
    plt.xlabel("Time Steps")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
#Assuming network is your trained network and time_steps is your input tensor
delta = 50.0  # Example increment value
n = 5  # Number of increments
increment_and_graph(network, time_steps, delta, n)
```
