```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations

def harmonic_wave_component(n, t):
    """
    Generates the n-th harmonic component of a square wave.
    
    Parameters:
    n (int): Harmonic number (must be odd).
    t (numpy array): Array of time intervals.

    Returns:
    numpy array: The n-th harmonic component values at the given time intervals.
    """
    return np.sin(n * t) / n

def generate_harmonic_combinations(x, t):
    """
    Generates all combinations of harmonic square wave components up to x harmonics.
    
    Parameters:
    x (int): Number of harmonics to include.
    t (numpy array): Array of time intervals.

    Returns:
    dict: A dictionary where keys are the number of harmonics in the combination and
          values are lists of numpy arrays representing the summed wave values for each combination.
    """
    harmonics = [harmonic_wave_component(n, t) for n in range(1, x * 2, 2)]
    harmonic_combinations = {}
    
    for i in range(1, x + 1):
        harmonic_combinations[i] = []
        for comb in combinations(harmonics, i):
            wave = np.abs(np.sum(comb, axis=0) * (4 / np.pi))
            harmonic_combinations[i].append(wave)
    
    return harmonic_combinations

# Example usage:
t = np.arange(0, 2 * np.pi, 0.01)  # Time intervals
x = 10  # Number of harmonics
harmonic_combinations = generate_harmonic_combinations(x, t)

# Plotting the waves
for num_harmonics, waves in harmonic_combinations.items():
    plt.figure(figsize=(10, 6))
    for wave in waves:
        plt.plot(t, wave)
    plt.title(f'Combinations of {num_harmonics} Harmonic(s)')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
```
