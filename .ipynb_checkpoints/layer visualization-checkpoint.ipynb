```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# SinCondense layer definition
class SinCondense(nn.Module):
    def __init__(self, input_dim, add_dim):
        super(SinCondense, self).__init__()
        assert input_dim % add_dim == 0, "input_dim must be divisible by add_dim"
        self.input_dim = input_dim
        self.add_dim = add_dim
        self.a = nn.Parameter(torch.randn(input_dim))
        self.phi = nn.Parameter(torch.randn(input_dim))

    def forward(self, t):
        # Ensure t has the correct shape
        if t.dim() == 1:
            t_expanded = t.unsqueeze(1).expand(-1, self.input_dim)
        else:
            t_expanded = t.expand(-1, self.input_dim)
        
        t_phi = t_expanded + self.phi
        sin_t_phi = torch.sin(t_phi)
        output = self.a * sin_t_phi
        output = output.view(-1, self.input_dim // self.add_dim, self.add_dim)
        output = output.sum(dim=-1)
        return output

    def count_params(self):
        return self.a.shape[0] + self.phi.shape[0]

# SinCondenseNetwork definition
class SinCondenseNetwork(nn.Module):
    def __init__(self, add_dim, depth):
        super(SinCondenseNetwork, self).__init__()
        layers = []
        input_dims = sorted([add_dim ** (i + 1) for i in range(depth)], reverse=True)
        self.input_shape = add_dim ** depth
        for input_dim in input_dims:
            layers.append(SinCondense(input_dim, add_dim))
            
        # Store the layers as a ModuleList
        self.layers = nn.ModuleList(layers)
        
    def forward(self, t):
        # Ensure the input t has the correct shape
        if t.dim() == 1:
            t = t.unsqueeze(-1).expand(-1, self.input_shape)
        
        outputs = []
        for layer in self.layers:
            t = layer(t)
            outputs.append(t)  # Capture the output of each layer
        return outputs
    
    def count_params(self):
        return sum(layer.count_params() for layer in self.layers)

# Initialize the SinCondenseNetwork
add_dim = 2
depth = 5
network = SinCondenseNetwork(add_dim, depth)

# Generate t values for plotting
data_t = torch.from_numpy(np.linspace(1, 10, 1000)).float()

# Forward pass through the network
with torch.no_grad():
    layer_outputs = network(data_t)

# Plot the output of each layer

for i, output in enumerate(layer_outputs):
    output_np = output.numpy()  # Convert to numpy for plotting
    plt.figure(figsize=(15, 6))
    plt.plot(data_t.numpy(), output_np, label=f'Layer {i+1} Output')

    plt.xlabel('t')
    plt.ylabel('Output')
    plt.title('SinCondenseNetwork Layer Outputs over t')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# Define the WeirdSinLayer class
class WeirdSinLayer(nn.Module):
    def __init__(self):
        super(WeirdSinLayer, self).__init__()
        
        self.param_A = nn.Parameter(torch.randn(1, 3))  # Amplitude
        self.param_phi = nn.Parameter(torch.randn(1, 3))  # Phase shift
        self.param_C = nn.Parameter(torch.randn(1, 3))  # Vertical shift

    def forward(self, t):
        # Ensure t has correct shape (batch_size, 1)
        if t.dim() == 1:
            t = t.unsqueeze(-1)
        
        # Expand params to match t's size (batch_size, 3)
        batch_size = t.shape[0]
        param_A = self.param_A.expand(batch_size, -1)
        param_phi = self.param_phi.expand(batch_size, -1)
        param_C = self.param_C.expand(batch_size, -1)
        
        # Calculate the sine wave parameters
        out_params = param_A * torch.sin(t + param_phi) #+ param_C
        
        out_A = out_params[:, 0].unsqueeze(-1)
        out_phi = out_params[:, 1].unsqueeze(-1)
        out_C = out_params[:, 2].unsqueeze(-1)
        
        # Final sine wave output
        out_t = out_A * torch.sin(t + out_phi) + out_C
        
        return out_t
    
class WeirdSinFinalLayer(nn.Module):
    def __init__(self):
        super(WeirdSinFinalLayer, self).__init__()
        
        self.param_A = nn.Parameter(torch.randn(1, 3))  # Amplitude
        self.param_phi = nn.Parameter(torch.randn(1, 3))  # Phase shift
        self.param_C = nn.Parameter(torch.randn(1, 3))  # Vertical shift

    def forward(self, t):
        # Ensure t has correct shape (batch_size, 1)
        if t.dim() == 1:
            t = t.unsqueeze(-1)
        
        # Expand params to match t's size (batch_size, 3)
        batch_size = t.shape[0]
        param_A = self.param_A.expand(batch_size, -1)
        param_phi = self.param_phi.expand(batch_size, -1)
        param_C = self.param_C.expand(batch_size, -1)
        
        # Calculate the sine wave parameters
        out_params = 1.0 * torch.sin(t + param_phi) #+ param_C
        
        out_A = out_params[:, 0].unsqueeze(-1)
        out_phi = out_params[:, 1].unsqueeze(-1)
        out_C = out_params[:, 2].unsqueeze(-1)
        
        # Final sine wave output
        out_t = out_A * torch.sin(t + out_phi) + out_C
        
        return out_t

# Define the WeirdSinNetworkInc class
class WeirdSinNetworkInc(nn.Module):
    def __init__(self, num_layers, inc=0.1):
        super(WeirdSinNetworkInc, self).__init__()
        
        # Dynamically create the specified number of WeirdSinLayer layers
        self.layers = nn.ModuleList([WeirdSinLayer() for _ in range(num_layers)])
        
        # Modify param_A in each layer based on the index and inc
        for idx, layer in enumerate(self.layers):
            additive = torch.tensor([10 + (idx * inc)]).unsqueeze(-1).float()
            layer.param_A = nn.Parameter(layer.param_A + additive)

    def forward(self, t):
        # Store outputs of each layer for visualization
        outputs = []
        for layer in self.layers:
            t = layer(t)
            outputs.append(t)
        return outputs

# Initialize WeirdSinNetworkInc with the desired number of layers
num_layers = 10
network = WeirdSinNetworkInc(num_layers, inc=0.1)

# Generate t values for plotting
data_t = torch.from_numpy(np.linspace(1, 100, 1000)).float()

# Forward pass through the network and get the output of each layer
with torch.no_grad():
    layer_outputs = network(data_t)

# Plot the output of each layer

for i, output in enumerate(layer_outputs):
    output_np = output.squeeze().numpy()  # Convert to numpy for plotting, remove extra dimensions
    plt.figure(figsize=(15, 6))
    plt.plot(data_t.numpy(), output_np, label=f'Layer {i+1} Output')

    plt.xlabel('t')
    plt.ylabel('Output')
    plt.title(f'WeirdSinNetworkInc Layer Outputs over t for layer {i + 1}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
class WeirdSinUnit(nn.Module):
    def __init__(self, num_layers, inc=0.1):
        super(WeirdSinUnit, self).__init__()
        
        # Create num_layers-1 WeirdSinLayer layers
        self.layers = nn.ModuleList([WeirdSinLayer() for _ in range(num_layers - 1)])
        
        # Add the WeirdSinFinalLayer as the last layer
        self.layers.append(WeirdSinFinalLayer())
        
        # Modify param_A in each WeirdSinLayer based on the index and inc
        for idx, layer in enumerate(self.layers[:-1]):  # Apply the increment only to WeirdSinLayer
            additive = torch.tensor([10 + (idx * inc)]).unsqueeze(-1).float()
            layer.param_A = nn.Parameter(layer.param_A + additive)

    def forward(self, t):
        # Store outputs of each layer for visualization
        outputs = []
        for layer in self.layers:
            t = layer(t)
            outputs.append(t)
        return outputs
    
num_layers = 5
network = WeirdSinUnit(num_layers, inc=0.1)

# Generate t values for plotting
data_t = torch.from_numpy(np.linspace(1, 100, 50)).float()

# Forward pass through the network and get the output of each layer
with torch.no_grad():
    layer_outputs = network(data_t)

# Plot the output of each layer

for i, output in enumerate(layer_outputs):
    output_np = output.squeeze().numpy()  # Convert to numpy for plotting, remove extra dimensions
    plt.figure(figsize=(15, 6))
    plt.plot(data_t.numpy(), output_np, label=f'Layer {i+1} Output')

    plt.xlabel('t')
    plt.ylabel('Output')
    plt.title(f'WeirdSinNetworkInc Layer Outputs over t for layer {i + 1}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
